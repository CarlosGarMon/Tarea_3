```python
entradas = { 1:"Ensalada verde", 2:"Totopos", 3:"Fruta picada" }
medio = { 1:"Crema de zanahoria", 2:"Sopa de tortilla", 3:"Pasta"}
fuerte = { 1:"Filete", 2:"Mojarra", 3:"Arrachera", 4:"Pechuga de pollo"}

print("------------- RESTAURANTE D' CARLOS -------------")
print("                Menú de temporada\n")
print("---> ENTRADA")
for i in entradas:
	print(f"{i}.- {entradas[i]}")
print("\n---> MEDIO")
for i in medio:
	print(f"{i}.- {medio[i]}")
print("\n---> PLATO FUERTE")
for i in fuerte:
	print(f"{i}.- {fuerte[i]}")

mesa = int(input ("\n           Introduce el numero de mesa: "))
print(f"\n           ----------- MESA {mesa} ----------- ")

orden = []
otra_vez = True
persona = 0
while otra_vez == True:
	individual = [mesa, persona + 1]
	print(f"\n------------------- COMENSAL {persona + 1}")
	validacion = False
	while validacion == False:
		opcion = int(input ("Introduzca el número de platillo para la entrada:  "))
		if (opcion in entradas.keys()):
			individual.append(entradas[opcion])
			validacion = True
		else:
			print("[ERROR: Verificar dato]")

	validacion = False
	while validacion == False:
		opcion = int(input ("Introduzca el número de platillo para el medio:  "))
		if (opcion in medio.keys()):
			individual.append(medio[opcion])
			validacion = True
		else:
			print("[ERROR: Verificar dato]")

	validacion = False
	while validacion == False:
		opcion = int(input ("Introduzca el número de platillo para el plato fuerte:  "))
		if (opcion in fuerte.keys()):
			individual.append(fuerte[opcion])
			validacion = True
		else:
			print("[ERROR: Verificar dato]")

	opcion = input ("Detalles o instrucciones adicionales: ")
	individual.append(opcion)
	orden.append(individual)

	validacion = False
	while validacion == False:
		opcion = int(input ("\nEscriba [1] para siguiente comensal | [0] para terminar el pedido \nOpcion: "))
		if opcion == 0 or opcion == 1:
			if opcion == 0:
				validacion = True
				otra_vez = False
			else:
				persona = persona + 1
				validacion = True
		else:
			print("[ERROR: Verificar dato]")

print("\n--------------- ORDEN REGISTRADA ---------------\nDetalle:\n")
for cliente in orden:
	print(f"\tMESA: { cliente[0] }")
	print(f"\tPERSONA: { cliente[1] }")
	print(f"\tENTRADA: { cliente[2] }")
	print(f"\tMEDIO: { cliente[3] }")
	print(f"\tFUERTE: { cliente[4] }")
	print(f"\tADICIONALES: { cliente[5] }")
	print("-------------------------------")

```
